```python
#1
def analyze_text(text):
    glasni = "аеиоуыэюя"
    unique_glasni = set()
    not_dyblicat = set()
    words = []
    clean = ""
    text_2 = text + " "
    for i in text_2:
        if i.isalpha():
            clean += i
            if i.lower() in glasni:
                unique_glasni.add(i.lower())
        else:
            if len(clean) > 0:
                if len(clean) >= 5:
                    first_char = clean[0].lower()
                    last_char = clean[-1].lower()
                    if first_char == last_char:
                        w = clean.lower()
                        if w not in not_dyblicat:
                            words.append(w)
                            not_dyblicat.add(w)
            clean = ""
    result = " ".join(words)
    return len(unique_glasni), result

q = "количество уникальных гласных, строка из найденных слов через пробеп"
s = analyze_text(q)
print(f"resalt: {s}")
```
