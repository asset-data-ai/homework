```python
with open("file.txt", "w") as f:
    for i in range(1, 11):
        f.write(str(i) + "\n")
with open("file.txt", "r") as f:
    print(f.read())
```

```python
q = ["asset", "eliya"]
with open("names.txt", "w") as f:
    for i in q:
        f.write(i + "\n")
with open("names.txt", "r") as f:
    for i in f:
        print(i.capitalize())
```

```python
q = [["asset", 18], ["eliya", 17]]
with open("names.csv", "w") as f:
    for i in q:
        f.write(f"{i[0]}, {i[1]}\n")
with open("names.csv", "r") as f:
    for i in f:
        print(i.capitalize())
```
