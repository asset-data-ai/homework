# Сохранение

# 1. Работа с .loc и .iloc (очень важно!)
iloc → как координаты (0,1)

loc → как Excel (по имени столбца)



#### .iloc — по индексам

```python
df.iloc[0]        # первая строка
df.iloc[0:3]      # строки 0-2
df.iloc[0, 1]     # строка 0, столбец 1
```

#### .loc — по названиям

```python
print(df.loc[0, "Country"])
print(df.loc[df["Popularity"] > 80, ["Country", "Popularity"]])
```

# Применение функций (apply)

```python
def grade_category(x):
    if x >= 90:
        return "A"
    elif x >= 75:
        return "B"
    else:
        return "C"

df["category"] = df["Popularity"].apply(grade_category)
print(df.head())
```

# Быстрые операции (без apply — лучше!)

```python
df["category"] = "C"
df.loc[df["Popularity"] >= 75, "category"] = "B"
df.loc[df["Popularity"] >= 90, "category"] = "A"
```

# GroupBy — самое важное
Идея:

“Разделили на группы → посчитали”

```python
df.groupby("Genre")["Popularity"].mean()
```

# или

```python
df.groupby("Artist").agg({
    "Popularity": "mean",
    "Awards_won": "sum"
})

# Перевод:

# для каждого артиста:

# средняя популярность
# сколько наград всего
```

# Pivot table (сводная)
Это просто “красивый groupby”

```python
df.pivot_table(
    values="Popularity",
    index="Genre",
    aggfunc="mean"
)
```

# Создание новых признаков (Feature Engineering)

```python
df["is_long"] = df["Duration_sec"] > 300 
print(df.head())
```

```python
###  Пример: популярность категориями
```

```python
df["pop_level"] = "low"
df.loc[df["Popularity"] > 50, "pop_level"] = "medium"
df.loc[df["Popularity"] > 80, "pop_level"] = "high" 
print(df.head())
```

# Бинирование (разбиение на группы)
Делим числа на категории

```python
df["pop_group"] = pd.cut(
    df["Popularity"],
    bins=[0, 30, 70, 100],
    labels=["low", "medium", "high"]
)
print(df.head())
#разбили популярность на интервалы как оценки (2,3,4,5)
```

# Работа с несколькими условиями (np.where) 

```python
import numpy as np

df["hit"] = np.where(
    (df["Popularity"] > 80) & (df["Streams_million"] > 500),
    "yes",
    "no"
)
```
