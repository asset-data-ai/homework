```python
def my_isalpha(q):
    if len(q) == 0:
        return False
    for i in q:
        if not ("a" <= i <= "z" or "A" <= i <= "Z"):
            return False
    return True

def my_isdigit(w):
    for i in w:
        if not ("0" <= i <= "9"):
            return False
    return True

def my_isupper(e):
    a = False
    for i in e:
        if "a" <= i <= "z":
            return False
        if "A" <= i <= "Z":
            a = True
    return a

def my_isspace(r):
    a = " \t\n\r\v\f"
    for i in r:
        if i not in a:
            return False
    return True

```

```python
#1
def analyze_text(text):
    q = "аеиоуыэюя"
    unique_glasni = set()
    not_dyblicat = set()
    words = []
    clean = ""
    text_2 = text + " "
    for i in text_2:
        if my_isalpha(i):
            clean += i
            if i.lower() in q:
                unique_glasni.add(i.lower())
        else:
            if len(clean) > 0:
                if len(clean) >= 5:
                    first_char = clean[0].lower()
                    last_char = clean[-1].lower()
                    if first_char == last_char:
                        w = clean.lower()
                        if w not in not_dyblicat:
                            words.append(w)
                            not_dyblicat.add(w)
            clean = ""
    result = " ".join(words)
    return len(unique_glasni), result


q = "количество уникальных гласных, строка из найденных слов через пробеп"
s = analyze_text(q)
print(f"resalt: {s}")
```

```python
#3
def top_k_words(text, k):
    clean = ""
    x = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    for i in text:
        if i in x:
            clean += " "
        else:
            clean += i.lower()
    q = {}
    for i in clean.split():
        if i in q:
            q[i] += 1
        else:
            q[i] = 1
    w = []
    for key, n in q.items():
        w.append((key, n))
    w.sort(key=lambda x: (-x[1], x[0]))
    result = []
    limit = min(k, len(w))
    for i in range(limit):
        result.append(w[i][0])
    return result

a = "apple go to: the apple@ garden next week, go"
s = 3
top_k_words(a, s)
```

```python
#4
def www(text):
    q = text.split()
    result = []
    for i in q:
        w = 0
        for j in i:
            if my_isupper(j):
                w += 1
        if w == 1:
            if not i[0].isupper() and not i[-1].isupper():
                result.append(i)
    return result
a = ("He's ranIng everY dAy ")
www(a)
```

```python
#5
def compress_text(text):
    result = ""
    count = 1
    first = text[0]
    for i in range(1, len(text)):
        current = text[i]
        if current.lower() == first.lower():
            count += 1
        else:
            result += first
            if count > 1:
                result += str(count)
            first = current
            count = 1
    result += first
    if count > 1:
        result += str(count)
    return result
a = "AaBbCddDee"
compress_text(a)
```

```python
#7
def palindrome_words(text):
    clean = ""
    x = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    for i in text:
        if i in x:
            clean += " "
        else:
            clean += i.lower()
    q = set(clean.lower().split())
    palindrome = []
    for i in q:
        if len(i) >= 3 and i == i[::-1]:
            palindrome.append(i)
    palindrome.sort(key=lambda x: (-len(x), x))
    return palindrome
a = "civic level. Racecar! I'm frty. wow, Pop."
palindrome_words(a)
```

```python
#9
def alternate_case_blocks(text, n):
    result = ""
    for i in range(0, len(text), n):
        block = text[i:i+n]
        block_number = i // n
        if block_number % 2 == 0:
            result += block.upper()
        else:
            result += block.lower()
    return result
a = "HelloWorldPython"
s = 3
alternate_case_blocks(a, s)
```

```python
#11
def common_unique_chars(s1, s2):
    s2_1 = set(s2)
    q = set()
    result = []
    for i in s1:
        if (not my_isdigit(i) and
            not my_isspace(i) and
            i in s2_1 and
            i not in q):
            result.append(i)
            q.add(i)
    return "".join(result)
w = "Hello World 23"
e = "Old 1 town"
common_unique_chars(w, e)
```

```python
#13
def replace_every_nth(text, n, char):
    result = list(text)
    q = set()
    i = 0
    while i < len(text):
        if my_isalpha(text[i]):
            start = i
            # ИСПРАВЛЕНИЕ 1: Выровняли отступы
            while i < len(text) and my_isalpha(text[i]):
                i += 1
            word_len = i - start
            if word_len < 3:
                for k in range(start, i):
                    q.add(k)
        else:
            i += 1

    for index, symbol in enumerate(result, 1):
        e = index - 1
        if index % n == 0:
            if symbol == " ":
                continue
            if my_isdigit(symbol):
                continue
            if e in q:
                continue
            result[e] = char

    return "".join(result)

# --- ТЕСТ ---
text = "Hi world 12345 Python is goood ok"
n_val = 2
char_val = "*"
print(replace_every_nth(text, n_val, char_val))
```

```python
#15
def word_pattern_sort(text):
    q = text.split()
    def w(e):
        return sum(1 for i in e.lower() if i in "aeiou")
    result = sorted(q, key=lambda e: (len(e), -w(e), e.lower()))
    return result
a = "apple cat onion bee bread idea milk"
word_pattern_sort(a)
```

```python
#16
def transform_list(nums):
    q = []
    for i in nums:
        if i < 0:
            continue
        if i % 2 == 0:
            q.append(i ** 2)
        else:
            if i > 10:
                w = sum(int(j) for j in str(i))
                q.append(w)
            else:
                q.append(i)
    return q
a = [-5, 4, 13, 7, 22, 99, -100, 3]
transform_list(a)
```

```python
#18
def flatten_and_filter(lst):
    result = []
    def q(w):
        for i in w:
            if isinstance(i, list):
                q(i)
            elif isinstance(i, int):
                if i >=10 and i % 4 != 0:
                    result.append(i)
    q(lst)
    return sorted(result)
a = [1, 5, [12, 22], [[-100, 15, 101]], [[[8, 44, 50]]]]
flatten_and_filter(a)
```
