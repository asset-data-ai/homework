```python
#1
unique_users = set()
user_spending = {}
total_purchases = 0
total_sum = 0

with open("shop_logs.txt", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        parts = line.split(";")
        user_id = parts[1]
        action = parts[2]
        unique_users.add(user_id)

       if action == "BUY":
            total_purchases += 1
            total_sum += int(parts[3])
            if user_id in user_spending:
                user_spending[user_id] += int(parts[3])
            else:
                user_spending[user_id] = int(parts[3])

if user_spending:
    best_buyer = max(user_spending, key=user_spending.get)
else:
    best_buyer = "none"

with open("report.txt", "w", encoding="utf-8") as f:
    f.write(f"Уникальных пользователей: {len(unique_users)}\n")
    f.write(f"Всего покупок: {total_purchases}\n")
    f.write(f"Общая сумма: {total_sum}\n")
    f.write(f"Самый активный покупатель: {best_buyer}\n")
    f.write(f"Средний чек: {total_sum / total_purchases:.2f}\n")
with open("report.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

```python
#2.2
import csv
all_employees = []
dept_salary = {}
with open("employees.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        salary = int(row["salary"])
        dept = row["department"]
        all_employees.append(row)
        row["salary"] = salary
        if dept not in dept_salary:
            dept_salary[dept] = []
        dept_salary[dept].append(salary)

total_salary = sum(i["salary"] for i in all_employees)
avg_total = total_salary / len(all_employees)

dept_averages = {}  #орташа dept
for dept, salary in dept_salary.items():
    avg = sum(salary) / len(salary)
    dept_averages[dept] = avg

best_dept_name = max(dept_averages, key=dept_averages.get)
best_dept_value = dept_averages[best_dept_name]
richest_emp = max(all_employees, key=lambda i: i["salary"])
print(f"Орташа жалақы: {avg_total}")
print(f"Ең бай бөлім: {best_dept_name} ({best_dept_value})")
print(f"Ең бай қызметкер: {richest_emp['name']}")


q = []
for i in all_employees:
    if i["salary"] > avg_total:
        q.append(i)

with open("high_salary.csv", "w", encoding="utf-8") as f:
    columns = ["name", "department", "salary"]
    writer = csv.DictWriter(f, fieldnames=columns)
    writer.writeheader()
    writer.writerows(high_earners)
print("\n")
with open("high_salary.csv", "r", encoding="utf-8") as f:
    q = csv.reader(f)
    for i in q:
        print(i)
```

```python
#3.2
import json
with open("orders.json", "r") as f:
        data = json.load(f)

total_revenue = 0
user_order_counts = {}
item_counts = {}
richest_order = None
max_price = 0
for i in data:
    total_revenue += i["total"]

    user = i["user"]    # читаем з к.п
    if user in user_order_counts:
        user_order_counts[user] += 1
    else:
        user_order_counts[user] = 1

    if i["total"] > max_price:
        max_price = i["total"]
        richest_order = i

    for j in i["items"]:
        if j in item_counts:
            item_counts[j] += 1
        else:
            item_counts[j] = 1
q = None
w = 0
for item, count in item_counts.items():
    if count > w:
        w = count
        q = item

top_user = richest_order["user"] if richest_order else None

resalt = {"total_revenue": total_revenue,
    "top_user": top_user,
    "most_popular_item": q,
    "total_orders": len(data)}
with open("summary.json", "w") as f:
    json.dump(resalt, f, indent=4)

with open("summary.json", "r") as f:
    print(f.read())
```

```python
#4.2
import csv
import json

suspicious_operations = []
suspicious_users = set()
spisok = {}
with open("transactions.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        user = row["user_id"]
        amount = int(row["amount"])

        if user in spisok:
            spisok[user] += 1
        else:
            spisok[user] = 1

        if amount > 500000:
            suspicious_operations.append(amount)
            suspicious_users.add(user)

for user, count in spisok.items():
    if count > 3:
        suspicious_users.add(user)

with open("fraud_report.txt", "w", encoding="utf-8") as f:
    f.write(f"Подозрительных транзакций: {len(suspicious_operations)}\n")
    f.write(f"Подозрительных пользователей: {len(list(suspicious_users))}\n")
    f.write(f"Список пользователей: {list(suspicious_users)}\n")
    f.write(f"Общая сумма подозрительных операций: {sum(suspicious_operations)}\n")

with open("fraud_users.json", "w") as f:
    json.dump({"fraud_users": list(suspicious_users)}, f, indent=4)

with open("fraud_report.txt", "r", encoding="utf-8") as f:
    print(f.read())
with open("fraud_users.json", "r", encoding="utf-8") as f:
    print(f.read())
```
