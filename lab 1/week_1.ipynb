```python
def my_isalpha(q):
    if len(q) == 0:
        return False
    for i in q:
        if not ("a" <= i <= "z" or "A" <= i <= "Z"):
            return False
    return True

def my_isdigit(w):
    for i in w:
        if not ("0" <= i <= "9"):
            return False
    return True

def my_isupper(e):
    a = False
    for i in e:
        if "a" <= i <= "z":
            return False
        if "A" <= i <= "Z":
            a = True
    return a

def my_isspace(r):
    a = " \t\n\r\v\f"
    for i in r:
        if i not in a:
            return False
    return True

def my_sorted(q):
    for i in range(len(q)):
        for j in range(1, len(q)):
            if q[j-1] >= q[j]:
                q[j-1], q[j] = q[j], q[j-1]
    return q

def my_isinstance(w, e):
    if type(e) == tuple:
        return type(w) in e
    return type(w) == e
```

```python
#1
def analyze_text(text):
    q = "аеиоуыэюя"
    unique_glasni = set()
    not_dyblicat = set()
    words = []
    clean = ""
    text_2 = text + " "
    for i in text_2:
        if my_isalpha(i):
            clean += i
            if i.lower() in q:
                unique_glasni.add(i.lower())
        else:
            if len(clean) > 0:
                if len(clean) >= 5:
                    first_char = clean[0].lower()
                    last_char = clean[-1].lower()
                    if first_char == last_char:
                        w = clean.lower()
                        if w not in not_dyblicat:
                            words.append(w)
                            not_dyblicat.add(w)
            clean = ""
    result = " ".join(words)
    return len(unique_glasni), result


q = "количество уникальных гласных, строка из найденных слов через пробеп"
s = analyze_text(q)
print(f"resalt: {s}")
```

```python
#3
def top_k_words(text, k):
    clean = ""
    x = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    for i in text:
        if i in x:
            clean += " "
        else:
            clean += i.lower()
    q = {}
    for i in clean.split():
        if i in q:
            q[i] += 1
        else:
            q[i] = 1
    w = []
    for key, n in q.items():
        w.append((key, n))
    w.sort(key=lambda x: (-x[1], x[0]))
    result = []
    limit = min(k, len(w))
    for i in range(limit):
        result.append(w[i][0])
    return result

a = "apple go to: the apple@ garden next week, go"
s = 3
top_k_words(a, s)
```

```python
#5
def compress_text(text):
    result = ""
    count = 1
    first = text[0]
    for i in range(1, len(text)):
        current = text[i]
        if current.lower() == first.lower():
            count += 1
        else:
            result += first
            if count > 1:
                result += str(count)
            first = current
            count = 1
    result += first
    if count > 1:
        result += str(count)
    return result
a = "AaBbCddDee"
compress_text(a)
```

```python
#7
def palindrome_words(text):
    clean = ""
    x = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    for i in text:
        if i in x:
            clean += " "
        else:
            clean += i.lower()
    q = set(clean.lower().split())
    palindrome = []
    for i in q:
        if len(i) >= 3 and i == i[::-1]:
            palindrome.append(i)
    palindrome.sort(key=lambda x: (-len(x), x))
    return palindrome
a = "civic level. Racecar! I'm frty. wow, Pop."
palindrome_words(a)
```

```python
#9
def alternate_case_blocks(text, n):
    result = ""
    for i in range(0, len(text), n):
        block = text[i:i+n]
        block_number = i // n
        if block_number % 2 == 0:
            result += block.upper()
        else:
            result += block.lower()
    return result
a = "HelloWorldPython"
s = 3
alternate_case_blocks(a, s)
```

```python
#13
def replace_every_nth(text, n, char):
    result = list(text)
    q = set()
    i = 0
    while i < len(text):
        if my_isalpha(text[i]):
            start = i
            # ИСПРАВЛЕНИЕ 1: Выровняли отступы
            while i < len(text) and my_isalpha(text[i]):
                i += 1
            word_len = i - start
            if word_len < 3:
                for k in range(start, i):
                    q.add(k)
        else:
            i += 1

    for index, symbol in enumerate(result, 1):
        e = index - 1
        if index % n == 0:
            if symbol == " ":
                continue
            if my_isdigit(symbol):
                continue
            if e in q:
                continue
            result[e] = char

    return "".join(result)

# --- ТЕСТ ---
text = "Hi world 12345 Python is goood ok"
n_val = 2
char_val = "*"
print(replace_every_nth(text, n_val, char_val))
```

```python
#20
def max_subarray_sum(nums, k):
    if len(nums) < k:
        return None
    q = 0
    w = 0
    e = None
    for i in range(k):
        r = nums[i]
        q += r
        if r <= 0:
            w += 1
    if w == 0:
        e = q
    for i in range(k, len(nums)):
        t = nums[i-k]
        q -= t
        if t <= 0:
            w -= 1
        y = nums[i]
        q += y
        if q <= 0:
            w += 1
        if w == 0:
            if e is None or q > e:
                e = q
    return e
a = [1, 2, 3, -5, 4, 5, 6, 0, 10]
max_subarray_sum(a, 3)
```

```python
#24
def longest_increasing_sublist(nums):
    q = []
    w = [nums[0]]
    for i in range(1, len(nums)):
        if nums[i] > nums[i-1]:
            w.append(nums[i])
        else:
            if len(w) > len(q):
                q = w
            w = [nums[i]]
    if len(w) > len(q):
        q = w
    return q
a = [1, 2, 5, 3, 4, 1, 7, 8, 9]
longest_increasing_sublist(a)
```

```python
#28
def moving_average(nums, k):
    q = []
    for i in range(len(nums) - k + 1):
        w = nums[i : i + k]
        if min(w) >= 0:
            e = sum(w) / k
            q.append(e)
    return q
a = [1, 2, 3, -4, 5, 6, 7]
moving_average(a, 3)
```

```python
#03
def merge_dicts_sum(d1, d2):
    q = d1.copy()
    for key, i in d2.items():
        if key in q:
            q[key] += i
        else:
            q[key] = i
    return q
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'b': 50, 'c': 100, 'd': 500}
merge_dicts_sum(d1, d2)
```

```python
#04
def filter_sets(sets_list):
    q = []
    for i in sets_list:
        if len(i) > 3 and min(i) >= 0:
            for j in i:
                if j % 2 == 0:
                    q.append(i)
                    break
    return q
a = [{1, 3, 5, 7}, {2, 4, 6, 8}, {10, -5, 20, 30}]
filter_sets(a)
```

```python
#06
def deep_sum(d):
    q = 0
    for i in d.values():
        if my_isinstance(i, dict):
            q += deep_sum(i)
        elif my_isinstance(i, list):
            for j in i:
                if my_isinstance(j, (int, float)):
                    q += j
        elif my_isinstance(i, (int, float)):
            q += i
    return q
a = {"a": 10, "b": [1, 2, 3], "c": {"x": 20, "y": [4, 5], "z": {"last": 100}, "d": "text"}}
deep_sum(a)
```

```python
#09
def common_elements_all(sets_list):
    if not sets_list:
        return set()
    q = sets_list[0].copy()
    for i in sets_list[1:]:
        q = q & i
        if not q:
            return set()
    return q
a = [{1, 2, 3, 4, 5}, {2, 3, 4, 10}, {3, 4, 11, 2}]
common_elements_all(a)
```

```python
#013
def invert_dict_strict(d):
    q = {}
    w = {}
    for i in d.values():
        q[i] = q.get(i, 0) + 1
    for key, i in d.items():
        if q[i] == 1:
            w[i] = key
    return w
a = {'a': 1, 'b': 2, 'c': 3, "d": 2}
print(invert_dict_strict(a))
```

```python
#014
def top_k_frequent(nums, k):
    q = {}
    for num in nums:
        if num in q:
            q[num] = q[num] + 1
        else:
            q[num] = 1
    w = list(q.items())
    w.sort(key=lambda x: (-x[1], x[0]))
    e = set()
    for i in range(len(w)):
        if i < k:
            e.add(w[i][0])
        else:
            break
    return e
a = [1, 1, 2, 2, 3, 100, 100]
s = 2
top_k_frequent(a, s)
```
