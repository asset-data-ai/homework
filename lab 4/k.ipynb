```python
#1-2
class Player:
    def __init__(self, id, name, hp):
        self._id = id
        self._name = name.strip().title()
        self._hp = max(0, hp)

    @classmethod #2
    def from_string(cls, data: str):
        parts = data.split(",")
        if len(parts) != 3:
            raise ValueError("Неверный формат строки. Ожидалось: 'id,name,hp'")
        try:
            id = int(parts[0].strip())
            name = parts[1].strip()
            hp = int(parts[2].strip())
            return cls(id, name, hp)
        except ValueError:
            raise ValueError("ID и HP должны быть целыми числами")

    def __str__(self):  #1
        return f"Player(id={self._id}, name='{self._name}', hp={self._hp})"

    def __del__(self): #1
        print(f"Player {self._name} удалён")
```

```python
#1
print("#1")
p = Player(1, " john ", 120)
print(p)

#2
print("==========================================")
print("#2")
p = Player.from_string("2, alice , 90")
print(p)
```

```python
#3
class Item:
    def __init__(self, id, name, power):
        self.id = id
        self.name = name.strip().title()
        self.power = power

    def __str__(self):
        return f"Item(id={self.id}, name='{self.name}', power={self.power})"

    def __eq__(self, other):
        if isinstance(other, Item):
            return (self.id, self.name, self.power) == (other.id, other.name, other.power)
        return False

    def __hash__(self):
        return hash((self.id, self.name, self.power))

i = Item(1, " Sword ", 50)
print(i)
print("--- Проверка работы с set ---")
sword1 = Item(1, "Sword", 50)
sword2 = Item(1, " sword ", 50)
shield = Item(2, "Shield", 20)
inventory = {sword1, sword2, shield}
for item in inventory:
        print(item)
```

```python
#4
class Inventory:
    def __init__(self):
        self._items: dict[int, Item] = {}

    def add_item(self, item: Item):
        if item.id not in self._items:
            self._items[item.id] = item

    def remove_item(self, id: int):
        self._items.pop(id, None)

    def get_items(self):
        return list(self._items.values())

    def unique_items(self):
        return set(self._items.values())

    def to_dict(self):
        return self._items.copy()

    #5
    def get_strong_items(self, min_power: int):
        is_strong = lambda item: item.power >= min_power
        return [item for item in self._items.values() if is_strong(item)]

sword1 = Item(1, "Sword", 50)
sword2 = Item(1, "Sword", 50)
shield = Item(2, "Shield", 20)
potion = Item(3, "Potion", 5)
inv = Inventory()
inv.add_item(sword1)
inv.add_item(sword2)
inv.add_item(shield)
inv.add_item(potion)
print("--- Список после добавления ---")
for item in inv.get_items():
    print(item)
inv.remove_item(2)
print("\n--- Словарь (to_dict) после удаления щита ---")
inventory_dict = inv.to_dict()
for item_id, item in inventory_dict.items():
    print(f"Key {item_id}: {item}")
print("\n--- Уникальные предметы (unique_items) ---")
fake_potion = Item(4, "Potion", 5)
inv.add_item(fake_potion)
print(f"Всего предметов в списке: {len(inv.get_items())}")
print(f"Уникальных предметов в set: {len(inv.unique_items())}")
```

```python
#6
from datetime import datetime
class Event:
    def __init__(self, type: str, data: dict):
        types = {"ATTACK", "HEAL", "LOOT"}
        if type not in types:
            raise ValueError(f"Неизвестный тип события: {type}. Доступны: {types}")
        self.type = type
        self.data = data
        self.timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    def __str__(self):
        return f"Event(type='{self.type}', data={self.data}, timestamp='{self.timestamp}')"

e1 = Event("ATTACK", {"damage": 20})
print(e1)
e2 = Event("HEAL", {"amount": 50, "source": "Potion"})
print(e2)
```
